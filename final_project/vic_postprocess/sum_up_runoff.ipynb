```python
import xarray as xr
```

```python
ds = xr.open_dataset("fluxes.1990-01-01.nc")
```

```python
ds
```

```python
# OUT_RUNOFF is surface runoff
# OUT_BASEFLOW is subsurface runoff
# total runoff = OUT_RUNOFF + OUT_BASEFLOW
ds_tot_runoff = ds["OUT_RUNOFF"] + ds["OUT_BASEFLOW"]
```

```python
# The unit for OUT_RUNOFF and OUT_BASEFLOW is mm/day
# we convert the unit to mm/sec
ds_tot_runoff_mms = ds_tot_runoff/86400
```

```python
# specify the unit in the attributes
ds_tot_runoff_mms.attrs['units'] = 'mm/s'
```

```python
# save the total runoff
ds_tot_runoff_mms.to_dataset(name='QRUNOFF').to_netcdf("runoff.1990-01-01.nc")
```
